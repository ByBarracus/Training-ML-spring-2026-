Задача: определить лучшего сотрудника в компании на основании имеющихся данных.

Среди данных, имеем:
1) Стаж в компании
2) Кол-во повышений, нормированное на 5 лет (если стаж 1 год то *5, так у Олега получилось 10, хотя по факту имеет лишь 2 повышения за год)
3) Кол-во лайков за лучшего сотрудника на портале

А также, мы попросили уполномоченных сотрудников, проголосовать за лучшего сотрудника и по их мнению, + разметить веса по параметрам (от наиболее важного к наименее важному)

Надо подобрать алгоритм, чтобы могли выйти на результат голосования, через первоначально имеющиеся данные.

Осмотрим данные:

```python
data.loc[4] = [0,4,'Оценка',0,0,0,0,0,0] #Обращаемся к 4 строке по индексу (еще не существует) и добавляем данные строкой
data
```

```python
del data['level_0'] #Удаляем level_0
data
```

Пошаговое объяснение команды
Эта строка кода выполняет присваивание новых заголовков столбцов для DataFrame df на основе данных из другого DataFrame data.

Разбор по частям
data.T — транспонирует DataFrame data, то есть меняет местами строки и столбцы. Теперь то, что было строками, стало столбцами, и наоборот.
.loc['Фактор'] — выбирает строку (в транспонированном DataFrame это была колонка) с индексом 'Фактор'. В результате получается серия (одномерный массив), где индексы — это старые имена столбцов, а значения — новые названия.
df.columns = ... — присваивает эти значения в качестве новых заголовков столбцов для DataFrame df.

```python
df = data.T # Транспонируем данные, строки = люди
df.columns = data.T.loc['Фактор']
coef_experts = df.loc['Веса'].iloc[:3] # Сохраним веса экспертов
df = df.drop(index=['Фактор', 'Веса'])
```

```python
df
```

Теперь построим линейную регресиию, в качестве известных x передадим известные нам фичи, в качестве y будут выступать результаты голосования уполномоченных сотрудников, тем самым попробуем подогнать данные под их экспертную оценку

```python
X = df[['Стаж','Кол-во повышений, нормированных на 5 лет','Кол-во «лайков» за лучшего сотрудника за год']]
y = df['Голосование']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, y)
print('R2', reg.score(X, y), 'coef', reg.coef_, 'bias', reg.intercept_) #Выведем R2, кэфы и биас для понимания (свободный член, для подгонки под результат более точно)
```

```python
df['pred'] = round((reg.coef_ * X).sum(axis=1) + reg.intercept_,1)
```

```python
df['pred']
```

```python
import numpy as np
df['pred'] = np.round(reg.predict(X), 1)
```

```python
df['pred']
```

Как видим, рассчитанные предикторы совпадают с экспертной оценкой.
Теперь, мы можем рассчитать эту метрику для любого сотрудника

Проверим, посчитаем оценку для второго Петра (с такими же данными 5,3,39) и второго Олега (10,2,19)

```python
# Ваши новые данные
N = np.array([[5, 3, 39]])  # Обратите внимание на двойные скобки!

# Получение прогноза
prediction = reg.predict(N)
print('Прогноз:', prediction)

```
